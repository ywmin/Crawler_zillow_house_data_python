```python
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd
from random import random
```

```python
# Build a zillow scraper
headers = {
    'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9',
    'accept-encoding': 'gzip, deflate, br',
    'accept-language': 'en-US,en;q=0.9',
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.0.0 Safari/537.36'
}
```

```python
urls = [f'https://www.zillow.com/charleston-sc/{i}_p/' for i in range(1, 15)]

data = {'address':[], 'price':[], 'house_details':[]}
session = requests.session()

for url in urls:

    resp = session.get(url, headers=headers)
    soup = BeautifulSoup(resp.content, 'html.parser')
    address = soup.find_all(class_="list-card-addr")
    price = soup.find_all(class_="list-card-price")
    house_details = soup.find_all(class_="list-card-details")
    for i in address:
        data['address'].append(i.text)
    for i in price:
        data['price'].append(i.text)
    for i in house_details:
        data['house_details'].append(i.text)
    time.sleep(1 + 2*random())

df = pd.DataFrame(data)
# print(data)   
```
